### Add noise - examples:

```python
from src.augmentation.add_noise import add_noise_from_file, add_gaussian_noise
```

* noise from file

```python
add_noise_from_file(audio_file_path="../../data/vox1_dev_wav/wav/id10001/1zcIwhmdeo4/00001.wav")
data_with_noise, _  = librosa.load(f"../../data/vox1_dev_wav/wav/id10001/1zcIwhmdeo4/00001_with_noise.wav",sr=None)

plt.figure(figsize=(9, 10))
plt.subplot(2,1,1)
plt.plot(data)
plt.title(f'audio')
plt.xlabel('Time')
plt.subplot(2,1,2)
plt.plot(data_with_noise)
plt.title(f'audio + noise from file')
plt.xlabel('Time')
```

* gaussian noise

```python
plt.figure(figsize=(9, 20))
plt.subplot(4,1,1)
plt.plot(data)
plt.title(f'audio')
plt.xlabel('Time')

i = 1
for loc, scale in zip([0, 0, 5], [1, 10, 1]):
    i+=1
    add_gaussian_noise(audio_file_path="../../data/vox1_dev_wav/wav/id10001/1zcIwhmdeo4/00001.wav", loc=loc, scale=scale)
    data_gaussian, _  = librosa.load(f"../../data/vox1_dev_wav/wav/id10001/1zcIwhmdeo4/00001_gaussian_noise_loc_{loc}_scale_{scale}.wav",sr=None)
    plt.subplot(4,1,i)
    plt.plot(data_gaussian)
    plt.title(f'audio + N({loc}, {scale})')
    plt.xlabel('Time')
```

### Subsampling - example

```python
from src.augmentation.subsampling import subsampling_audio
```

```python
plt.figure(figsize=(9, 20))
plt.subplot(4,1,1)
plt.plot(data)
plt.title(f'audio')
plt.xlabel('Time')

i = 1
for interval in [2, 5, 10]:
    i+=1
    subsampling_audio(audio_file_path="../../data/vox1_dev_wav/wav/id10001/1zcIwhmdeo4/00001.wav", interval=interval)
    data_subsampling, _  = librosa.load(f"../../data/vox1_dev_wav/wav/id10001/1zcIwhmdeo4/00001_subsampling_{interval}.wav",sr=None)
    plt.subplot(4,1,i)
    plt.plot(data_subsampling)
    plt.title(f'interval={interval}')
    plt.xlabel('Time')
plt.show()
```
